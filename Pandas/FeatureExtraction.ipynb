## Make a new Column for episode count

```python
df.loc[2]["Title"]
```

```python
def extract_episodes(txt):
    check = False
    data = ""
    
    for i in txt:
        if i == ")":
            break
        if i == '(':
            check = True
        if check == True and i != '(':
            data += i
    ans = data.replace(" eps", "")
    return int(ans)
```

```python
df["Episodes"] = df["Title"].apply(extract_episodes)
```

## Make a new Column for time stamp

```python
df.loc[44]["Title"]
```

```python
def extract_timeStamp(txt):
    for i in range(0, len(txt)):
        if txt[i] == ')':
            return txt[i+1 : i+20]

    return ""
```

```python
df["Timestamps"] = df["Title"].apply(extract_timeStamp)
```

## Which is the Longest running anime

```python
def calculate_run_duration(ts):
    start_str, end_str = ts.split(' - ')
    start = pd.to_datetime(start_str.strip())
    end = pd.to_datetime(end_str.strip())
        
    return (end - start).days

df['Run_Duration'] = df['Timestamps'].apply(calculate_run_duration)
```
