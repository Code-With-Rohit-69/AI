# Merging, joining, Concatation

### Merge 2 DataFrames

```python
emp_data = {
    'Emp_ID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'City': ['New York', 'London', 'Paris', None, 'Mumbai']
}

salary_data = {
    'Emp_ID': [101, 102, 103, 104, 105],
    'Salary': [70000, 85000, np.nan, 62000, 58000],
    'Dept': ['IT', 'HR', 'IT', 'Finance', 'Marketing']
}

df_employee = pd.DataFrame(emp_data)
df_salary = pd.DataFrame(salary_data)
```

```python
df_employee
```

```python
df_salary
```

```python
pd.merge(df_employee, df_salary, on="Emp_ID", how="inner")
```

```python
pd.merge(df_employee, df_salary, on="Emp_ID", how="outer")
```

```python
pd.concat([df_employee, df_salary])
```

# Group by Aggregation

```python
dates = pd.date_range(start='2024-01-01', periods=8)

data = {
    'Category': ['Electronics', 'Furniture', 'Electronics', 'Groceries', 
                 'Furniture', 'Electronics', 'Groceries', 'Electronics'],
    'Store': ['North', 'South', 'North', 'West', 'East', 'South', 'West', 'North'],
    'Sales': [1200.50, 450.00, np.nan, 150.75, 2100.00, 300.00, 85.00, 500.25],
    'Quantity': [3, 1, 2, 10, 5, 2, np.nan, 4],
    'Date': dates,
}

df = pd.DataFrame(data)
```

```python
df
```

```python
df.groupby("Category")["Sales"].sum()
```

```python
df.groupby("Store")["Sales"].sum()
```

### Group by multiple columns

```python
df.groupby(["Category", "Store"])["Sales"].sum()
```

### Aggregation

```python
df.groupby("Store")["Sales"].mean()
```

# 🥅 Pivot Tables

```python
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South'] * 5,
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice'] * 5
}

df = pd.DataFrame(data)

df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)
```

```python
df
```

```python
pd.pivot_table(df, values="Sales", index="Region", columns="Product")
```

```python
pivot = pd.pivot_table(df, values=["Sales", "Units"], index="Region", columns="Product")
```

```python
pivot
```

```python
pivot.fillna(0)
```

```python
pivot
```

### Cross Tables

```python
pd.crosstab(df["Region"], df["Product"])
```
