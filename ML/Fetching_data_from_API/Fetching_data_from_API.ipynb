```python
dfs = []

for i in range(1, 429):
    response = requests.get(
        f"https://api.themoviedb.org/3/movie/top_rated?api_key=YOUR_API_KEY&language=en-US&page={i}"
    )

    data = response.json()

    if "results" in data:
        temp_df = pd.DataFrame(data["results"])
        temp_df = temp_df[["id","title","overview","release_date","popularity"]]
        dfs.append(temp_df)
    else:
        print(f"Error on page {i}:", data)

new_df = pd.concat(dfs, ignore_index=True)
```
