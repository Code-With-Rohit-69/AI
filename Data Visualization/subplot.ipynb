```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.randint(1, 20, size=5)
y = np.random.randint(1, 20, size=5)

plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title("Line Plot")

plt.subplot(1, 2, 2)
plt.bar(x, y, width=.8)
plt.title("Bar Plot")

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

x = np.random.randint(1, 20, size=5)
y = np.random.randint(1, 20, size=5)

ax[0].plot(x, y, color="red", marker="o")
ax[0].set_title("1st Plot")
ax[0].grid()

ax[1].bar(x, y, color="skyblue", width=1.9)
ax[1].set_title("2nd Plot")

fig.suptitle("Multiple Plots")

plt.tight_layout()

plt.show()
```
