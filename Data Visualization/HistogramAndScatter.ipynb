```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
nums = np.random.randint(100, 1000, size=10)
plt.hist(nums, bins=10, color="orange", edgecolor="black")

plt.title("Histogram")
plt.grid(False)
plt.show()
         
```

```python
x = np.random.randint(10, 100, size=10)
y = np.random.randint(10, 100, size=10)

plt.scatter(x, y, marker="x", color="red")

for xi, yi in zip(x, y):
    plt.annotate(f"({xi}, {yi})", (xi, yi), textcoords="offset points",
                 xytext=(5, 5))

plt.grid(True)
plt.title('Scatter Chart with Annotations')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```
