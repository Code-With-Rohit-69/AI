```python
import numpy as np

x = np.arange(1, 50, 3)
y = x * 3

plt.plot(x, y, color="blue", marker="o")
plt.title("Simple Line Plot with Labels")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y, color="green", marker="o", linestyle="-", label="Data Points", linewidth="1")
plt.grid(True)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot with Markers")
plt.legend()

plt.show()
```

```python
# Annotations

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.figure(figsize=(8, 6))

plt.plot(x, y, marker="o", linestyle="-.", color="red")

for xi, yi in zip(x, y):
    plt.annotate(f"({xi}, {yi})",
                 (xi, yi),
                 textcoords="offset points",
                 xytext=(0, 10),
                 ha="center"
                )

plt.title('Line Chart with Annotations')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

plt.show()
```

```python
# Multiple Plots on the Same Axis
import numpy as np

x = np.arange(1, 10, 1)
y = 2 * x

x1 = np.arange(2, 11, 2)
y1 = np.arange(3, 12, 2)

plt.figure(figsize=(10, 4))

plt.plot(x, y, color="green", linestyle="-", label="y = 2x")
plt.plot(x1, y1, color="red", linestyle="-.", label="Second Line")

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.grid(True)
plt.legend()
plt.show()
```

```python
# Multiple Plots on the Same Axis
import numpy as np

x = np.array([1, 2, 3, 4])
y = x * 2
y1 = [3, 5, 7, 9]

plt.figure(figsize=(8, 4))

plt.plot(x, y, color="green", linestyle="-", label="y = 2x")
plt.plot(x, y1, color="red", linestyle="-.", label="Second Line")
plt.fill_between(x, y, y1, color="lightblue")

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Filled Area Between Two Lines")

plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.rand(5)
y = np.random.rand(5) 

plt.figure(figsize=(4, 4))

for xi, yi in zip(x, y):
    plt.annotate(f"({xi}, {yi})", (xi, yi), xytext=(5, 5), textcoords='offset points')

plt.scatter(x, y, color="purple", marker="x")
plt.title("Scatter Plot Example")   
plt.grid(True, linestyle='-.')
plt.show()
```
